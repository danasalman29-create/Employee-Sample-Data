```python
import pandas as pd
```

### LOAD DATA

```python
df=pd.read_excel("Employee Sample Data - A.xlsx")
```

```python
print(df.to_string())
```

```python
df.info()
```

### CLEAN DATA

```python
df.duplicated()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Is_Exited'] = df['Exit Date'].notnull().astype(int)
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Annual Salary']=df['Annual Salary'].fillna(df['Annual Salary'].median())
df['Bonus %']=df['Bonus %'].fillna(df['Bonus %'].median())
```

```python
df['Age'].median()
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Country'] = df['Country'].fillna(df['Country'].mode()[0])

df['Department'] = df['Department'].fillna("Unknown")
df['City'] = df['City'].fillna("Unknown")
df['Ethnicity'] = df['Ethnicity'].fillna("Unknown")
df['Job Title'] = df['Job Title'].fillna("Unknown")

df['Hire Date'] = df['Hire Date'].fillna(df['Hire Date'].median())
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Full Name'])
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Exit Date'])
```

```python
df.isnull().sum()
```
